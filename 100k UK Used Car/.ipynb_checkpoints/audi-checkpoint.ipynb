## Feature Engineering

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ctransformer = ColumnTransformer([("encoded_data", OneHotEncoder(sparse=False), [0,3,5]),])
ct_encoded_results = ctransformer.fit_transform(dataset)

# Get Feature Names of Encoded columns
ctransformer.get_feature_names()

# Converting the numpy array into a pandas dataframe
df_ct_encoded_data = pd.DataFrame(ct_encoded_results, columns=ctransformer.get_feature_names())

# Viewing first few rows of data
df_ct_encoded_data.head()
```

```python
# Dropping dummy variables to avoid multicollinearity
df_ct_encoded_data.drop(['encoded_data__x0_ A1', 'encoded_data__x1_Automatic', 'encoded_data__x2_Diesel'], inplace=True, axis=1)

# Viewing few rows of data after dropping dummy varibles
df_ct_encoded_data.head()
```

```python
# Concatenating the encoded dataframe with the original dataframe
df = pd.concat([dataset.reset_index(drop=True), df_ct_encoded_data.reset_index(drop=True)], axis=1)

# Dropping 'model', 'transmission', 'fuelType' columns as they are encoded
df.drop(['model', 'transmission', 'fuelType'], inplace=True, axis=1)

# Viewing few rows of data
df.head()
```

```python
def plotting_3_chart(df, feature):
    ## Importing seaborn, matplotlab and scipy modules. 
    import seaborn as sns
    import matplotlib.pyplot as plt
    import matplotlib.gridspec as gridspec
    from scipy import stats
    import matplotlib.style as style
    style.use('fivethirtyeight')

    ## Creating a customized chart. and giving in figsize and everything. 
    fig = plt.figure(constrained_layout=True, figsize=(12,8))
    ## creating a grid of 3 cols and 3 rows. 
    grid = gridspec.GridSpec(ncols=3, nrows=3, figure=fig)
    #gs = fig3.add_gridspec(3, 3)

    ## Customizing the histogram grid. 
    ax1 = fig.add_subplot(grid[0, :2])
    ## Set the title. 
    ax1.set_title('Histogram')
    ## plot the histogram. 
    sns.distplot(df.loc[:,feature], norm_hist=True, ax = ax1)

    # customizing the QQ_plot. 
    ax2 = fig.add_subplot(grid[1, :2])
    ## Set the title. 
    ax2.set_title('QQ_plot')
    ## Plotting the QQ_Plot. 
    stats.probplot(df.loc[:,feature], plot = ax2)

    ## Customizing the Box Plot. 
    ax3 = fig.add_subplot(grid[:, 2])
    ## Set title. 
    ax3.set_title('Box Plot')
    ## Plotting the box plot. 
    sns.boxplot(df.loc[:,feature], orient='v', ax = ax3 );
```

```python
plotting_3_chart(df, 'price')
```

## Splitting the dataset into the Training set and Test set

```python
X = df.drop(['price'], axis=1)
y = df['price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Select models

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.linear_model import Ridge,ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

models = []
models.append(('LR', LinearRegression()))
# models.append(('PNR', PolynomialFeatures(degree = 4))) pnr has no score model
models.append(('pipelineSVR', Pipeline([('scalar1',StandardScaler()),
                     ('pca1',PCA(n_components=2)),
                     ('lr_classifier',SVR(kernel='linear'))])))
models.append(('Ridge', Ridge(alpha=2,max_iter=1000,random_state=0)))
models.append(('RFR', RandomForestRegressor(n_estimators = 10, random_state = 0)))
models.append(('GBR', GradientBoostingRegressor(random_state=0)))
models.append(('XGBReg', XGBRegressor(random_state=0)))
models.append(('DCR', DecisionTreeRegressor(random_state = 0)))
```

## ## Select best of the best model

```python
# coefficient of determination R^2 of the prediction
for name, model in models:
    model.fit(X_train, y_train)
    scores = model.score(X_train,y_train)
    print(name,"Score: {:.2f} %".format(scores*100))
#           ,"Standard Deviation: {:.2f} %".format(scores.std()*100))
```

```python
## Highest score for training = DCR, highest score for test = RFR

# regressor = DecisionTreeRegressor(random_state = 0)
# regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor = XGBRegressor(random_state=0)
regressor.fit(X_train, y_train)
```

```python
results = X_test.copy()
results
```

```python
y_pred = regressor.predict(X_test)

results["predicted"] = y_pred 
results["actual"] = y_test

results = results[['predicted', 'actual']]
results['predicted'] = results['predicted'].round(2)
results
```

```python
from sklearn.metrics import r2_score

score = regressor.score(X_test,y_test)
r2score = r2_score(y_test, y_pred)

print("Score: {:.2f} %".format(score*100))
print("R2-Score: {:.2f} %".format(r2score*100))
```
