## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Importing dataset

```python
dataset = pd.read_csv('dataR2.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.isnull().sum() # Checking & finding the missing values of dataset 
```

## Splitting dataset into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.31, random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#print(X_train)
```

```python
#print(X_test)
```

## Training "Decision Tree" model on the Training set

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score, KFold

#for name, model in models:
#     kFold = KFold(n_splits=10)
accuracies = cross_val_score(classifier, X = X_train, y = y_train, cv = 10, scoring='accuracy')
print("Accuracy: {:.2f} %".format(accuracies.mean()*100), 
      "Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)

print('y_pred: ' + '\n', y_pred)
print('\n')

#reshape from "1 x length of column" to "length of column x 1" 
print('y_pred+y_test: '+'\n', np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),axis=1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
print(cm)
print(cr)
```
